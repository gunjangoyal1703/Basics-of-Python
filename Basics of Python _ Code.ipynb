```python
# Take marks from 5 students, find sum and average
m1 = float(input("Enter marks for student 1: "))
m2 = float(input("Enter marks for student 2: "))
m3 = float(input("Enter marks for student 3: "))
m4 = float(input("Enter marks for student 4: "))
m5 = float(input("Enter marks for student 5: "))

total_sum = m1 + m2 + m3 + m4 + m5
average = total_sum / 5

print("Total Sum:", total_sum)
print("Average:", average)
```

```python
# Write a Python program to find the area of a circle.
r = float(input("Enter radius: "))
area = 3.14 * r * r
print("Area of circle =", area)
```

```python
# Find area and perimeter of a rectangle
length = float(input("Enter the length of the rectangle: "))
width = float(input("Enter the width of the rectangle: "))

area = length * width
perimeter = 2 * (length + width)

print(f"Area: {area}")
print(f"Perimeter: {perimeter}")
```
