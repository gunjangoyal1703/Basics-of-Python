```python
# Access Items
SubjectList = ["AMT", "DCLD", "EDA", "PES", "CT", "TPCS", "IKS"]
print(SubjectList[0])
print(SubjectList[1])
print(SubjectList[2])
```

```python
# Reverse/Negative Indexing
SubjectList = ["AMT", "DCLD", "EDA", "PES", "CT", "TPCS", "IKS"]
print(SubjectList[-1])
print(SubjectList[-2])
print(SubjectList[-3])
```

```python
# Slicing
SubjectList = ["AMT", "DCLD", "EDA", "PES", "CT", "TPCS", "IKS"]
print(SubjectList[1:3]) # 1 is inclusive, 3 is exclusive
print(SubjectList[ :5]) # Nothing is written, assumes as 0 (First 4 entries)
print(SubjectList[3:]) # Nothing is written, assumes last entry
print(SubjectList[-3:]) # Will access lat three entries (Last 3 entries)
print(SubjectList[2:-2]) # Negative indexing
print(SubjectList[0:5]) # First 4 entries
print(SubjectList[-4:-2]) # Access PES and CT (using negative indexing)
print(SubjectList[-4:]) # Access from PES to end
```

```python
# Change Item Value
SubjectList = ["AMT", "DCLD", "EDA", "PES", "CT", "TPCS", "IKS"]
SubjectList[6] = "ECSF"
print(SubjectList)
```

```python
# Append Items
SubjectList = ["AMT", "DCLD", "EDA", "PES", "CT", "TPCS", "IKS"]
SubjectList.append("ECSF") # Adds one more entry at the end
print(SubjectList)
```

```python
# Insert Items
SubjectList = ["AMT", "DCLD", "EDA", "PES", "CT", "TPCS", "IKS"]
SubjectList.insert(0, "ECSF") # Adds entry at mentioned position
print(SubjectList)
```

```python
# Extend List
NameList1 = ["Anand", "Bhavya", "Cathy"]
NameList2 = ["Deepak", "Eminem", "Farrah"]
NameList1.extend(NameList2) # List 2 will be added to List 1
print(NameList1)
```

```python
#Slicing Examples
Numbers = [10, 20, 30, 40, 50]
print("Elements from index 1 to 3:") ; print(Numbers[1:4])
print("First three elements:") ; print(Numbers[:4])
print("Elements from index 2 onwards:") ; print(Numbers[-3:])
print("All elements:") ; print(Numbers[:])
```

```python
# Various Marks
Marks = [78, 85, 90, 66, 88]
print("Maximum Marks:",max(Marks))
print("Minimum Marks:",min(Marks))
print("Total Marks:",sum(Marks))
print("Number of Students:",len(Marks))
print("Average Marks:",sum(Marks)/len(Marks))
print("Sort Marks in Ascending Order:",Marks.sort())
print(Marks) # We need to print again for sort as the original gets edited
```

```python
# Grocery Shopping List
print("GroceryList:"); GroceryList = ["Salt","Eggs","Pineapple","Milk","Curd","Oil"]; print(GroceryList)
print("Adding Forgotten Items:"); GroceryList.append("Chocolates"); print(GroceryList)
print("Removing Purchased Items:"); GroceryList.remove("Oil"); print(GroceryList)
print("Final List:"); print(GroceryList)
```

```python
# List of Phone Contacts
print("ContactList:"); ContactList = ["Ram","Raman","Geeta","Nita","Tina","Prashant"]; print(ContactList)
print("Adding New Contact:"); ContactList.append("Diviya"); print(ContactList)
print("Deleting Existing Contsct:"); ContactList.remove("Nita"); print(ContactList)
print("Sort Contacts in Alphabetical Order:",ContactList.sort()); print(ContactList)
```

```python
# Slicing Examples
List1 = [20,30,50,60,40,10,70]
print(List1[1:3])
print(List1[1:3:1])
print(List1[1:5:2])
print(List1[1:4:3])
print(List1[4:6:-1]) # Goes one step back
print(List1[6:4:-1])
print(List1[6:4:-2])
print(List1[-2:-4:-2])
```

```python
# Temperature Analysis
PuneTemp = [34, 32 ,37, 43, 23, 34, 43, 25, 36,38]
print("Pune's Temperature in first 5 days:",(PuneTemp)[:5])
print("Pune's Temperature in last 5 days:",(PuneTemp)[5:])
print("Pune's Temperature:",(PuneTemp))
print("Highest Temperature:",max(PuneTemp))
print("Lowest Temperature:",min(PuneTemp))
print("Average Temperature:",sum(PuneTemp)/len(PuneTemp))
```
